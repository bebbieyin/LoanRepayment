This notebook is for preparing the data for training. 

## Problem Statement & Objective

Typically after a person applies for a loan, an underwriting process undergoes to determine whether to approve or reject the loan. The borrower's documents and details are looked through to assess the risk of whether they will able to pay the debt off. Underwriting can either be performed in an automated or manual way. Automated means to use pre-defined rules or a ML model, manually is through human judgement. 

But even after going through the rigorous process of underwriting, there's a number of approved loans are being defaulted, causing companies to lose money. Therefore, it is important to further improve the underwriting process. Instead of using pre-defined rules or human judgement, a model is develop to determine the probabilty of whether a person is able to repay their loan. 

## Proposed solution

For this task, the model will be predicting two classes, one for the person is able to repay the loan, another for that person is not able to repay. It will be a binary classification problem. The **loanStatus** in loan.csv will be used as the target variable. 

As the loanStatus columns contains different type of values, preprocessing has to be done to convert them into two classes.

```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
df_loan = pd.read_csv(r"data/loan.csv")
df_payment = pd.read_csv(r"data/payment.csv")
df_underwrite = pd.read_csv(r"data/clarity_underwriting_variables.csv")
```

```python
# get the values inside loanStatus
print(df_loan.loanStatus.value_counts())
# visualize
df_loan.loanStatus.value_counts().plot(kind='bar')
```

For the postive class, we will get the rows where the debt is successfully paid off:
- Paid Off Loan
- Settlement Paid Off

For the negative class, we will get the following rows where the loan is defaulted:
- Settled Bankruptcy
- Charged Off Paid Off
- Charged Off

We only choose the values that can indicate whether the customer has the ability to pay of the loan or not. Other status like cancelled or voided with vague meaning will not be included.

```python
# encode the labels to yes or no (0 or 1)

dataset = df_loan.loc[df_loan['loanStatus'].isin(['Paid Off Loan','Settlement Paid Off','Settled Bankruptcy','Charged Off Paid Off','Charged Off'])]
dataset['loanStatus'] = dataset.loanStatus.replace({'Paid Off Loan': 0, 'Settlement Paid Off': 0})
dataset['loanStatus'] = dataset.loanStatus.replace({'Settled Bankruptcy': 1, 'Charged Off Paid Off': 1, 'Charged Off': 1})
```

Create new feature. The number of loans that the client has applied for. 

```python
df_loan_num = df_loan.anon_ssn.value_counts().rename_axis('anon_ssn').to_frame('No. of loans requested')
```

```python
# add in the new features (no. of loans requested and the clear fraud score)

dataset = pd.merge(dataset, df_loan_num, on='anon_ssn',how='left')
dataset = pd.merge(dataset, df_underwrite[['clearfraudscore','underwritingid']], left_on='clarityFraudId',right_on='underwritingid',how='left')
```

```python
# drop NaN rows

dataset = dataset.dropna(axis=0)
```

```python
dataset.describe()
```

From the table above, we can see that the columns isFunded and hasCF has a standard deviation of 0. That means that all the data for that column is the same, it does not help the model learn any new meaningful information, therefore we will add it to the features to be removed. Other features that will be dropped are unnecessary ones like loanid, ssn number and so on.  

```python
# save a version of the dataset for data analysis

dataset.to_csv(r'data/eda_data.csv',index=False)
```

```python
# drop the unnecessary columns like ID

unwanted_cols = ['loanId','anon_ssn','applicationDate','originatedDate','originated','clarityFraudId','underwritingid','approved',
                 'isFunded','hasCF']
dataset= dataset[dataset.columns.drop(unwanted_cols)]
```

Next step is to encode the features with categorical variables to a numerical label, as the model can't process text directly. 

```python
# convert the columns with categorical values to numerical class with label encoder

cat_features = ['payFrequency','state','leadType','fpStatus']
dataset = dataset.copy()
for i in cat_features:
    dataset[i] = dataset[i].astype('category')
    dataset[i] = dataset[i].cat.codes
```

```python
# export dataset to csv

dataset.to_csv(r'data/mydata.csv',index=False)
```
