## Understanding Data & Domain Knowledge

- **Underwriting** : Process that lenders take to determine the risk of approving a loan for the borrower. Typically includes verifying income, assets, debt, credit score, etc.
- **APR (Annual Percentage Rate)** : Annual cost of a loan to a borrower, similar to interest rate but include other fees and charges.

- **Origination** : First step of loan process. Obtain all the client's data and documents. Difference of this with underwriting is that in origination, only we are only obtaining the documents of the clients, underwriting requires having someone/something to analyze client's application and determine if client is qualified for the role.

- **ACH(Automated Clearing House)**: Electronic fund transfer made between banks and credit unions. Used for all kinds of fund transfer transactions, including direct deposit of paychecks and monthly debits for routine payment

### Average clear fraud score for rejected and approved applications

```python
df_clarity = pd.merge(df_loan, df_underwrite[['clearfraudscore','underwritingid']], left_on='clarityFraudId',right_on='underwritingid',how='inner')
df_rejected = df_clarity.loc[df_clarity['approved'] ==False]
df_approved = df_clarity.loc[df_clarity['approved'] ==True]
```

```python
# calc the median and mean of the rejected and approved applications

print("Median of approved: %f"%(df_approved['clearfraudscore'].median()))
print("Mean of approved: %f"%(df_approved['clearfraudscore'].mean()))
print("Median of rejected: %f"%(df_rejected['clearfraudscore'].median()))
print("Mean of rejected: %f"%(df_rejected['clearfraudscore'].mean()))
```

Strangely, the rejected applications has a higher clear fraud score. Median is also used in addition to mean as it is more prone to outliers.

### Threshold for approving loan based on clear fraud score

```python
# visualize if the data contains a lot of outliers

sns.boxplot(x=df_clarity['clearfraudscore'])
```

```python
def find_outliers_IQR(df):

    
    q1=df.quantile(0.25)

    q3=df.quantile(0.75)

    IQR=q3-q1

    outliers = df[((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]

    return outliers

min_score = find_outliers_IQR(df_clarity['clearfraudscore']).min()
max_score = find_outliers_IQR(df_clarity['clearfraudscore']).max()

print("Outliers of clear fraud score typically lies between %d and %d"%(min_score,max_score))
```

The above box plot calculates outliers based on the InterQuartile Range (IQR) rule. 
The average clear fraud score is between 600-800, outliers values have a lower clear fraud score ranging from 100 to 300. There is a hug gap between the the average and outlier values. <br>

One of the rules that we can set for the system is to automatically reject applications with low clear fraud score, after that human underwriters will only need to reviewed those with higher fraud score, reducing the workload. From the analysis above, we can set the threshold to 600. Applications with clear fraud score will pass automated underwriting test and be accessed by humans while those under 600 will be automatically rejected. 

## Inferential Statistics Analysis

Perform some inferential analysis and make some assumptions about the data. <br>
The dataset used here is the one that is preprocessed in notebook 1, the samples here are all the applications that have been approved. <br>
For all the following analysis, we will use alpha = 0.05 to compare with p-value.

```python
alpha = 0.05
```

#### Research Question: Does the clear fraud score has an effect on the loan amount that is being approved? Does higher clear fraud score means we can approve loan with higher amount and vice versa?
Null hypothesis : The clear fraud score and loan amount has no significant relationship <br>
Alternative hypothesis : There is a significant relationship between the clear fraud score and loan amount.

Use z-test to determine if there is a relationship between the two numerical variables. 

```python

tstat,pvalue = ztest(dataset['clearfraudscore'],dataset['loanAmount'])

if pvalue < alpha:
    print(
      'p value less than alpha. Reject null hypothesis.\n' 
        'There is a relation between clear fraud score and loan amount.'
  )
else:
    print(
      'p value more than alpha. Accept null hypothesis.\n'
    'Clear fraud score and loan amount are independent of each other'

  )
```

Pearson correlation to further determine the relationship between the two

```python
corr, pvalue = pearsonr(np.array(dataset['clearfraudscore']), np.array(dataset['loanAmount']))
print('Pearsons correlation score: %.3f' % corr)
plt.scatter(np.array(dataset['clearfraudscore']), np.array(dataset['loanAmount']))
```

There exists a positive correlation between the two variables. <br>
That means when one variable increases, another increases as well. <br>
The higher the clear fraud score, the higher the loan amount. 

#### Research Question: Does the number of times that customers have to make payment effects their ability to pay off loans?

Null hypothesis : The payment frequency and loan status has no significant relationship <br>
Alternative hypothesis : There is a significant relationship between the payment frequency and loan status.

Chi squared test to explore the relationship between two categorical variables

```python
# create contigency table for payFrequency vs loanStatus
info = pd.crosstab(index=dataset['payFrequency'], columns=dataset['loanStatus'])

# perform chi square
chi2, p, dof, _ = chi2_contingency(info.values.tolist())

#print(info.values.tolist())
#print('chi2 value: ', chi2)
#print('p value: ', p)
#print('dof value: ', dof)


if p < alpha:
    print(
      'p value less than alpha. Reject null hypothesis.\n' 
        'There is a relation between payment frequency and loan status.'
  )
else:
    print(
      'p value more than alpha. Accept null hypothesis.\n'
    'Payment frequency and loan status are independent of each other'

  )
```
